#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Carregar a base de dados
df_yahoo = pd.read_csv('./dadosLimpos.csv', header=[0, 1], index_col=[0])

# Definir uma janela de 90 dias
janela = 90

# Selecionar o fechamento ajustado ('Adj Close') dos ativos AAPL, AMZN, GOOGL, MSFT, NVDA
ativos_adj_close = df_yahoo['Adj Close'][['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'NVDA']]

# Calcular a média móvel de 90 dias
media_movel_90 = ativos_adj_close.rolling(window=janela).mean()

# Calcular o desvio padrão em uma janela de 90 dias
desvio_padrao_90 = ativos_adj_close.rolling(window=janela).std()

# Criar gráfico de linhas para a média móvel dos ativos
fig_media_movel = px.line(media_movel_90, title='Média Móvel de 90 dias (Fechamento Ajustado)')
fig_media_movel.show()

# Criar gráfico de linhas para o desvio padrão dos ativos
fig_desvio_padrao = px.line(desvio_padrao_90, title='Desvio Padrão de 90 dias (Fechamento Ajustado)')
fig_desvio_padrao.show()
````

o grafico da media móvel conseguimos analisar o desempenho de preços de ativos. enquanto o desvio padrão indica que o ativo tem uma grande variação de preços.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
lista_indicadores = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['AAPL']*len(lista_indicadores)
colunas = list(zip(lista_indicadores, ativo))

msft = df_yahoo.loc['14/08/2024':, colunas]
msft.columns = lista_indicadores

msft.head()
````

````python
graph = {
    'x': msft.index,
    'open': msft.Open,
    'close': msft.Close,
    'high': msft.High,
    'low': msft.Low,
    'type': 'candlestick',
    'name': 'AAPL',
    'showlegend': True
}
````

````python
grafico = graph_objects.Figure(
    data=[graph],
    layout_title="Índice Microsoft",
    # layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
